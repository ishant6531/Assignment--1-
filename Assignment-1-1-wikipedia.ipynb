# Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
#importing  the required librares
from bs4 import BeautifulSoup
import requests
```

```python
# Now we will send the request to the webpage server to get the source code of the page which is to be scraped by using the request library

page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
```

```python
# Let's see show the webpage source code downloaded 
page.content
```

```python
# Now we will parse this source code by using BeautifulSoup and the parser used here is html.parser as the source code is the HTML.
soup=BeautifulSoup(page.content,'html.parser')
```

```python
#Lets see how our parsed structure look like and we will use prettify function to present tit in more understandable and readable format
print(soup.prettify())
```

```python
# Finding all headers
titles = soup.find_all(['h1', 'h2','h3','h4','h5','h6'])
```

```python
titles
```

```python
#Printing all headers.
print('List all the header tags :', *titles, sep='\n\n')
```
